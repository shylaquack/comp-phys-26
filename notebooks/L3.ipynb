```python
#photons hitting some rectangle, want to generate random 2d coords for where a photon could hit
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
```

```python
from scipy.stats.distributions import norm

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def pdf(x, mu=10, sigma=2.):
    return 1/np.sqrt(2*np.pi*sigma**2) * np.exp(-(x - mu)**2/(2*sigma**2))
```

```python
mu = 10.
sigma = 2.

xs = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)

plt.plot(xs, pdf(xs), ls='--')
plt.ylim(ymin=0)

plt.ylabel(r'$p(x|\mu={}, \sigma={})$'.format(mu, sigma))
plt.xlabel('$x$');
```

```python
N = 100
samples =norm.rvs(loc=mu,scale=sigma,size=100000000)
plt.hist(samples,density=True,bins=50);
```

```python
def plot_dist(N=100,mu=10,sigma=2):
    xs= np.linspace(mu-5*sigma,mu +5*sigma,100)
    plt.plot(xs,norm.pdf(xs,loc=mu,scale = sigma))
    samples = norm.rvs(loc=mu,scale=sigma,size=N)
    plt.hist(samples,bins=30,alpha=0.5,density=True)

    plt.label(r'$p(x|\mu={},\sigma={})$'.format(mu,sigma))
    plt.ylim(0,.3)
```

```python
interactive(plot.dist,N=100)
```

```python
#Monte Carlo Simulation
#coin flips
true_bias = np.random.rand()
true_bias
```

```python
def flip_coin(nflips = 20, coin_bias = true_bias):
    #obs = np.zeros(nflips, dtype=bool)
    random_inputs = np.random.rand(nflips)
    #below bias gives one value, above give another
    obs = random_inputs < coin_bias
    return obs

flip_coin(int(1e6))
#estimate bias 
np.mean(flip_coin(int(1e6)))
```
